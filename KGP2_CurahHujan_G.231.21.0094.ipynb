```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Membaca dataset
dataset = pd.read_csv('CurahHujan.csv')
```

```python
dataset
```

```python
# Data curah hujan
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
curah_hujan = [
    [489, 482, 422],
    [308, 420, 410],
    [299, 420, 378],
    [260, 340, 283],
    [199, 140, 125],
    [147, 128, 105],
    [40, 59, 50],
    [70, 60, 2],
    [135, 60, 60],
    [230, 402, 338],
    [420, 498, 640],
    [400, 405, 425]
]

# Menentukan posisi bar
posisi_bulan = np.arange(len(bulan))
width = 0.2

# Membuat bar chart
plt.bar(posisi_bulan, [curah[0] for curah in curah_hujan], width, color='blue', label='1981-1990')
plt.bar(posisi_bulan + width, [curah[1] for curah in curah_hujan], width, color='red', label='1991-2000')
plt.bar(posisi_bulan + width * 2, [curah[2] for curah in curah_hujan], width, color='green', label='2001-2010')

# Mengatur label sumbu x dan y, serta judul
plt.xlabel('Bulan', weight='bold')
plt.ylabel('Curah Hujan (mm)', weight='bold')
plt.title('Perbedaan Warna Setiap Tahun', weight='bold')

# Mengatur posisi dan label pada sumbu x
plt.xticks(posisi_bulan + width, bulan)

# Menambahkan garis-garis horizontal kecil di belakang
for i in range(20, 700, 20):
    plt.axhline(y=i, color='gray', linestyle='-', alpha=0.3)

# Menampilkan legenda
plt.legend(loc="upper center", ncol=3, bbox_to_anchor=(0.5, 1.2))

# Menampilkan plot
plt.show()
```
